```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

random.seed(10)
```

```python
data = pd.read_csv('Prepared_data.csv')
data = data.drop(data.columns[0], axis = 1)
data.head()
```

```python
# Assingning variables
features = data.drop(['Tilt_angle'], axis = 1)
target = data['Tilt_angle']
features.shape,target.shape
```

```python
features
```

```python
# Splitting data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, target,
                                                                            test_size = 0.4,random_state=True,shuffle=False)

print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
train_features
```

```python
# plt.xticks(rotation=45)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))

train_features = scaler.fit_transform(train_features)
test_features = scaler.fit_transform(test_features)
```

```python
train_features
```

# KNN

```python
# Import the model we are using
from sklearn.neighbors import KNeighborsRegressor

# Instantiate model 
knn_ = KNeighborsRegressor(n_neighbors=8,p=2)

# Train the model on training data
knn_.fit(train_features, train_labels)


# Use the forest's predict method on the test data
predictions_ts_knn = knn_.predict(test_features)
predictions_tr_knn = knn_.predict(train_features)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from math import sqrt

# Calculate the absolute errors
MAE_ts_knn = mean_absolute_error(test_labels,predictions_ts_knn)
MSE_ts_knn = mean_squared_error(test_labels,predictions_ts_knn)
RMSE_ts_knn = sqrt(mean_squared_error(test_labels,predictions_ts_knn))
r2_ts_knn = r2_score(test_labels,predictions_ts_knn)



# Calculate the absolute errors
MAE_tr_knn = mean_absolute_error(train_labels,predictions_tr_knn)
MSE_tr_knn = mean_squared_error(train_labels,predictions_tr_knn)
RMSE_tr_knn = sqrt(mean_squared_error(train_labels,predictions_tr_knn))
r2_tr_knn = r2_score(train_labels,predictions_tr_knn)


# Print out the mean absolute error (mae)
print('KNN - Mean Absolute Error-test data :', round(MAE_ts_knn, 3))
print('KNN - Mean Square Error-test data:', round(MSE_ts_knn, 3))
print('KNN - Root Mean Square Error-test data:', round(RMSE_ts_knn, 3))
print('KNN - r2-test data:', round(r2_ts_knn, 3))

# Print out the mean absolute error (mae)
print('KNN - Mean Absolute Error-train data:', round(MAE_tr_knn, 3))
print('KNN - Mean Square Error-train data:', round(MSE_tr_knn, 3))
print('KNN - Root Mean Square-Error train data:', round(RMSE_tr_knn, 3))
print('KNN - r2-train data:', round(r2_tr_knn, 3))
```

# SVM

```python
from sklearn.svm import SVR

SVM_r = SVR(kernel='rbf')
SVM_r.fit(train_features, train_labels)

# Use the forest's predict method on the test data
predictions_ts_SVR = SVM_r.predict(test_features)
predictions_tr_SVR = SVM_r.predict(train_features)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from math import sqrt

# Calculate the absolute errors
MAE_ts_SVR = mean_absolute_error(test_labels,predictions_ts_SVR)
MSE_ts_SVR = mean_squared_error(test_labels,predictions_ts_SVR)
RMSE_ts_SVR = sqrt(mean_squared_error(test_labels,predictions_ts_SVR))
r2_ts_SVR = r2_score(test_labels,predictions_ts_SVR)



# Calculate the absolute errors
MAE_tr_SVR = mean_absolute_error(train_labels,predictions_tr_SVR)
MSE_tr_SVR = mean_squared_error(train_labels,predictions_tr_SVR)
RMSE_tr_SVR = sqrt(mean_squared_error(train_labels,predictions_tr_SVR))
r2_tr_SVR = r2_score(train_labels,predictions_tr_SVR)


# Print out the mean absolute error (mae)
print('SVR - Mean Absolute Error-test data :', round(MAE_ts_SVR, 3))
print('SVR - Mean Square Error-test data:', round(MSE_ts_SVR, 3))
print('SVR - Root Mean Square Error-test data:', round(RMSE_ts_SVR, 3))
print('SVR - r2-test data:', round(r2_ts_SVR, 3))

# Print out the mean absolute error (mae)
print('SVR - Mean Absolute Error-train data:', round(MAE_tr_SVR, 3))
print('SVR - Mean Square Error-train data:', round(MSE_tr_SVR, 3))
print('SVR - Root Mean Square-Error train data:', round(RMSE_tr_SVR, 3))
print('SVR - r2-train data:', round(r2_tr_SVR, 3))
```
